```python
#12 Создать класс линии 
class Line(object):
    def __init__(self, x1, y1, x2, y2):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
        self.get_coords()
    
    def get_coords(self):
        self.k = (self.y2 - self.y1) / (self.x2 - self.x1)
        self.b = self.y2 - self.k * self.x2
        
    def print_screen(self):
        for line in self.sp:
            for symbol in line:
                print(symbol, end='')
            print()
        
    def draw(self):
        self.sp = [['.']*40 for i in range(40)]
        self.sp[self.x1][self.y1] = '*'
        self.sp[self.x2][self.y2] = '*'
        for i in range(len(self.sp)):
            for j in range(len(self.sp[i])):
                if j == round(self.k * i + self.b) and self.x1 <= i <= self.x2 and self.y1 <= j <= self.y2:
                    self.sp[i][j] = '*'
        self.print_screen()

ln = Line(1, 5, 7, 10)
ln.draw()
```

```python
#16 Создать несколько объектов класса треугольник, прямоугольник и фигура, поместить их в список. Расчитать суммарную площадь всех форм, находящихся в списке.

class Rectangle(Figure):
    def __init__(self, a, b):
        self.a, self.b = a, b
 
    def get_area(self):
        m = self.a * self.b
        return m         

class Triangle(Figure):
    def __init__(self, a1, b1, c1):
        self.a1, self.b1, self.c1 = a1, b1, c1
    def get_area(self):
        p = (self.a1+self.b1+self.c1)/2
        n = (p * (p-self.a1) * (p-self.b1) * (p-self.c1))**(1/2)
        return n
        

class Fig(Figure):    
    def __init__(self, x, y, w, l):
        self.x = x
        self.y = y
        self.w = w
        self.l = l
        
    def get_area(self):                   
        z = (self.w-self.x) * (self.l-self.y)
        return z

       
 
r = Rectangle(20, 40)
tr = Triangle(30, 40, 60)
f = Fig(1, 1, 20, 20)
A = []
A.extend([r.get_area(), tr.get_area(), f.get_area()])
print('суммарная площадь:', float('{:.3f}'.format(sum(A))))
```
